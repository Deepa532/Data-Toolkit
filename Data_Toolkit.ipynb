21. Twenty First Question
-> NumPy’s array slicing is faster than Python’s list slicing because:

Memory Efficiency:
NumPy arrays store data in contiguous memory blocks and return views (not copies) when sliced, avoiding extra memory allocation.

Low-level Optimizations:
NumPy operations are implemented in C, which performs slicing at a much lower level compared to Python’s dynamic list objects.

Fixed Data Types:
NumPy arrays have homogeneous data types, making element access and slicing faster than Python lists, which store heterogeneous objects.
